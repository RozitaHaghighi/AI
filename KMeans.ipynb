```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error
from PIL import Image
import os

# Function to compress an image using K-means clustering with a predefined K value and calculate MSE
def compress_image_with_k_and_mse(image_path, k):
    original_image = Image.open(image_path).convert('RGB')
    original_pixels = np.array(original_image).reshape(-1, 3)

    # Apply K-means clustering with the predefined K value
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0)
    kmeans.fit(original_pixels)
    compressed_pixels = kmeans.cluster_centers_[kmeans.labels_]
    compressed_pixels = np.clip(compressed_pixels.astype('uint8'), 0, 255)
    compressed_image_np = compressed_pixels.reshape(original_image.size[1], original_image.size[0], 3)
    compressed_image = Image.fromarray(compressed_image_np)

    # Calculate MSE for the compressed image
    mse = mean_squared_error(original_pixels, compressed_pixels)

    # Save the compressed image
    dirname, filename = os.path.split(image_path)
    basename, ext = os.path.splitext(filename)
    compressed_image_file = os.path.join(dirname, f"{basename}_compressed_k{k}.png")
    compressed_image.save(compressed_image_file, format='PNG')

    return compressed_image_file, mse

# Load the image and define the path
image_path = '/content/bird.png'  # Update the path if necessary
k = 20  # Predefined K value (you can adjust this as needed)

# Compress the image with the predefined K value and calculate MSE
compressed_image_path, mse = compress_image_with_k_and_mse(image_path, k)

# Load the original and compressed image for display
original_image = Image.open(image_path)
compressed_image = Image.open(compressed_image_path)

# Display the results
print(f"Compressed image saved at: {compressed_image_path}")
print(f"MSE for K = {k}: {mse}")

# Display the original and compressed images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(np.array(original_image))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f'Compressed Image with K = {k}')
plt.imshow(np.array(compressed_image))
plt.axis('off')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error
from PIL import Image
import os

# Function to compress an image using K-means clustering and calculate MSE for each K
def compress_image_kmeans(image_path, k_values):
    original_image = Image.open(image_path).convert('RGB')
    original_pixels = np.array(original_image).reshape(-1, 3)

    best_k = None
    best_mse = float('inf')
    best_compressed_image_file = ""
    mse_values = {}  # Dictionary to store MSE for each K

    # Iterate over k_values to find the best k
    for k in k_values:
        # Apply K-means clustering to the image
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=0)
        kmeans.fit(original_pixels)
        compressed_pixels = kmeans.cluster_centers_[kmeans.labels_]
        compressed_pixels = np.clip(compressed_pixels.astype('uint8'), 0, 255)
        compressed_image_np = compressed_pixels.reshape(original_image.size[1], original_image.size[0], 3)
        compressed_image = Image.fromarray(compressed_image_np)

        # Calculate MSE for the compressed image
        mse = mean_squared_error(original_pixels, compressed_pixels)
        mse_values[k] = mse  # Store the MSE for this K

        # Check if this is the best k so far
        if mse < best_mse:
            best_mse = mse
            best_k = k
            # Save the best compressed image
            dirname, filename = os.path.split(image_path)
            basename, ext = os.path.splitext(filename)
            best_compressed_image_file = os.path.join(dirname, f"{basename}_compressed_k{k}.png")
            compressed_image.save(best_compressed_image_file, format='PNG')

    return best_k, best_mse, best_compressed_image_file, mse_values

# Load the image and define the path
image_path = '/content/bird.png'  # Update the path if necessary
k_values = [5, 10, 15, 20, 30, 40, 50, 64]

# Find the best K and compress the image
best_k, best_mse, best_compressed_image_path, mse_values = compress_image_kmeans(image_path, k_values)

# Display the MSE values for all K values
print("MSE values for each K:")
for k, mse in mse_values.items():
    print(f"K = {k}: MSE = {mse}")

# Load the original and best compressed image for display
original_image = Image.open(image_path)
compressed_image = Image.open(best_compressed_image_path)

# Display the best results
print(f"\nBest k: {best_k}")
print(f"Best MSE: {best_mse}")
print(f"Compressed image saved at: {best_compressed_image_path}")

# Display the original and compressed images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(np.array(original_image))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f'Best_Compressed Image with K = {best_k}')
plt.imshow(np.array(compressed_image))
plt.axis('off')

plt.show()
```
